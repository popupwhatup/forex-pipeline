```python
def generate_trade():
    instruments = ['EURUSD', 'XAUUSD', 'GBPUSD', 'USDJPY', 'AUDUSD']
    directions = ['BUY', 'SELL']

    return {
        'instrument': random.choice(instruments),
        'direction': random.choice(directions),
        'profit': round(random.uniform(-500, 1500), 2),
        'trade_date': datetime.now() - timedelta(days=random.randint(0, 30))
    }
print(generate_trade())

```

```python
cursor = conn.cursor()

for i in range(20):
    trade = generate_trade()
    cursor.execute("""
        INSERT INTO trades (instrument, direction, profit, trade_date)
        VALUES (%s, %s, %s, %s)
    """, (trade['instrument'], trade['direction'], trade['profit'], trade['trade_date']))

conn.commit()
print("Inserted 20 trades!")
```
